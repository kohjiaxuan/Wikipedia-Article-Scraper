```python
#THE FARM, GETTING THE FOOD/DATA FROM A SITE using requests
import requests
page = requests.get('https://en.wikipedia.org/wiki/Trance_music')
contents = page.content

#MAKE A STEW, BEAUTIFULSOUP, IT INDENTS ALL THE HTML BY PARSING IT AND ARRANGING IT IN READABLE FORMAT
from bs4 import BeautifulSoup
soup = BeautifulSoup(contents, 'html.parser')
# soup.find_all('a') works because all links have a href

#find all paragraphs or <p>/<li> with the relevant text
trancetext = soup.find_all('p')
trancetext2 = soup.find_all('li')
# type(trancetext)

#get paragraphs from trancetext with special format into a list
para=[]
xx = []
for paragraph in trancetext: #append paragraphs starting with <p>
    para.append(paragraph)
    xx.append(paragraph)

for paragraph in trancetext2: #append paragraphs starting with <li>
    para.append(paragraph)
    xx.append(paragraph)
    
    
#REASON WHY WE HAVE TO DO TWO FOR LOOPS WITH TWO TRANCETEXT IS BECAUSE THE FIND_ALL FOR ARRAY IS NOT IN ORDER

#COMMENCE CLEANING OF NONSENSE HTML <> and WIKI LINK [no]

para = list(str(para)) #chop everything into letters for usage

start = 0 #is letter currently inside tag <>
end = 0 #has <> just ended, need to check for , if it just ended to not copy a comma after <>
first = 1 #first letter is [, need to omit
bracket = 0 #check if letter is inside bracket
li = 0 #check for <li> to line break
p = 0 #check for <p> to line break
para2 = []
for letter in para:
    if first == 0:
        if letter == '<': #tells python to stop reading letters inside a bracket
            start = 1
        elif letter == '>': #next letter can be read since its out of bracket, unless its another <
            start = 0
            end = 1
        elif end == 1 and letter == ',': #skip COMMA reading when it occurs like </p>, at end of para
            end = 0
            continue
        elif letter == '[':
            bracket = 1
            end = 0
        elif letter == ']':
            bracket = 0
            end = 0
        elif start == 0 and bracket == 0: #ALL CLEAR TO READ LETTER
            end = 0
            para2.append(letter)
    if letter == '<':
        li = 1
    elif letter != 'l' and li == 1:
        li = 0
    elif letter == 'l' and li == 1:
        li = 2
    elif letter == 'i' and li == 2:
         li = 3
    elif letter == '>' and li == 3:
        para2.append('\n\n')
        li = 0
    if letter == '<':
        p = 1
    elif letter != 'p' and p == 1:
        p = 0
    elif letter == 'p' and p == 1:
        p = 2
    elif letter == '>' and p == 2:
        para2.append('\n')
        p = 0
    first = 0 #Had an issue with the first letter being [, after skipping this, the [number] checks can run

para2=''.join(para2) #combine back all letters and spaces
print(para2)

#REMOVE UNWANTED ARRAYS
# para1 = []
```

```python
#COUNT TOTAL WORDS AND COMMON WORDS

from collections import Counter

para3 = para2.split()
for index, word in enumerate(para3):
    para3[index] = word.lower()

trancecounter = Counter(para3)


#convert counter to list to dictionary then sum up total word count using for loop in word[key]
word = dict(trancecounter.most_common())
top200tranceword = trancecounter.most_common(250)

trancelist = []
count = 0
for key in word:
    count += word[key]
    if count < 2500:
        trancelist.append(key)
print(trancelist)
print('\n')
print('Total word count is ' + str(count))
```
