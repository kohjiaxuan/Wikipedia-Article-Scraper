```python
#WEBSITE:
#https://www.analyticsvidhya.com/blog/2015/10/beginner-guide-web-scraping-beautiful-soup-python/

#import the library used to query a website
import urllib.request as urllib #if you are using python3+ version, import urllib.request
#specify the url
wiki = "https://en.wikipedia.org/wiki/Most_common_words_in_English"
#Query the website and return the html to the variable 'page'
page = urllib.urlopen(wiki) #For python 3 use urllib.request.urlopen(wiki)
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page, "lxml")
print(soup)
```

```python
soup.title.string #TITLE OF WEBSITE SOUP.TITLE.STRING
```

```python
all_links = soup.find_all("a") #ALL HTML TAGS STARTING WITH <A, E.G. <A HREF, <A TITLE AND FULL PARAGRAPH

wiktwords = []
for link in all_links: #FOR LOOP CAN GET INTO THE ELEMENTS OF THE SOUP AND GET ITEMS ENCLOSED INSIDE USING .GET("NAME")
        wiktwords.append(link.get("title")) #TAG STARTING WITH A, CONTENT ENCLOSED INSIDE TITLE=""
        #print(link.get("title")) #shows list of items appended, common words all start with wikt:

wordcount = 0        
top100 = []
for words in wiktwords:
    words2 = str(words)
    if words2.find("wikt:") != -1:
        tempsliceword = list(str(words2))
        newword = []
        x = 0
        for letter in tempsliceword:
            if x > 4:
                newword.append(letter)
            x += 1
        cleanword = ''.join(newword)
        print(cleanword)
        top100.append(cleanword)
        wordcount += 1
print('\nTotal word count is ' + str(wordcount)+'\n\n')

print(top100)
        
        

    
```
