```python
#WEBSITE:
#https://www.analyticsvidhya.com/blog/2015/10/beginner-guide-web-scraping-beautiful-soup-python/

#import the library used to query a website
import urllib.request as urllib #if you are using python3+ version, import urllib.request
#specify the url
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
#Query the website and return the html to the variable 'page'
page = urllib.urlopen(wiki) #For python 3 use urllib.request.urlopen(wiki)
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page, "lxml")
print(soup)
```

```python
soup.title.string #TITLE OF WEBSITE SOUP.TITLE.STRING
```

```python
all_links=soup.find_all("a") #BS4.ELEMENT.RESULTSET
for link in all_links:
    print (link.get("href")) #THERE IS SPECIAL GET FUNCTION IN BEAUTIFUL SOUP TO GET HREF LINK
```

```python
# ONLY METHOD TO GET ELEMENTS INSIDE BODY CLASS TAG
body_class=soup.find_all("body")
for line in body_class:
    print (line.get("class"))
```

```python
# ONLY METHOD TO GET ELEMENTS INSIDE DIV ID TAG
div_id=soup.find_all("div")
for line in div_id:
    print (line.get("id"))
```

```python
right_table=soup.find('table', {'class':'wikitable sortable plainrowheaders'}) 
#identifying html code with table, <table class = 'wikitable sortable plainrowheaders'>

#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

#HTML KNOWLEDGE
#<th>	Defines a header cell in a table
#<tr>	Defines a row in a table
#<td>	Defines a cell in a table

#FOR LOOP FOR EACH ROW (<TR>), FIND THE CELLS (<TD>) AND THE HEADER CELLS (<TR>)

#findAll tag is case sensitive
for row in right_table.findAll("tr"): #find all entries with <tr> tag and put them in a list, then get each of them as row
    cells = row.findAll('td') #further search for td tag within each ROW
    headercell=row.findAll('th') #To store second column data, further search for th tag
#     print('Cell variable')
#     print(cells) - find out how variable looks like
#     print('-------')
#     print('States variable')
#    print(states) - find out how variable looks like
#     print(len(states))
    if len(cells)==5: #Only extract table body not heading, table body have 5 elements per cells <td> within <tr> tag
        A.append(cells[0].find(text=True)) #find text in between tags in each element to append
        B.append(headercell[1].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))        
        G.append(headercell[0].find(text=True))


alphabet = {0:A, 1:B, 2:C, 3:D, 4:E, 5:F, 6:G}

for index in alphabet:
    print(index)
    print(alphabet[index])
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(G,columns=['Number']) #6, header cell
df['State/UT']=B #1, header cell
df['Admin_Capital']=A #0, normal cell
df['Legislative_Capital']=C #2, normal cell
df['Judiciary_Capital']=D #3, normal cell
df['Year_Capital']=E #4, normal cell
df['Former_Capital']=F #5, normal cell

df
```
